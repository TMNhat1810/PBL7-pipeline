```python
import pandas as pd
```

```python
all_data = pd.read_csv('data/news.csv')
```

```python
current = pd.read_csv('cache/current.csv')
data = all_data.merge(current, how='left', on=['Date', 'Url'], indicator=True)
```

```python
data = data[data['_merge'] == 'left_only'].reset_index()
```

```python
data
```

```python
from transformers import AutoTokenizer
from transformers import TFAutoModelForSeq2SeqLM
from transformers import pipeline

checkpoint = "./model/model_finetuned"
tokenizer = AutoTokenizer.from_pretrained("VietAI/vit5-base")
model = TFAutoModelForSeq2SeqLM.from_pretrained(checkpoint, max_length=64, min_length=20)
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)
```

```python
def summarizer_wrapper(text):
    result = summarizer(text)
    return result[0]['summary_text']
```

```python
from_index = 0
to_index = None
```

```python
def batch_predict(from_index=0, to_index=1000):
    if to_index != None:
        if from_index >= to_index: return []
    
    bp = data[['Categorical', 'Date', 'Url']][from_index:to_index]
    bp['Pred'] = data['Content'][from_index:to_index].apply(summarizer_wrapper)
    
    return bp
```

```python
preds = batch_predict(from_index, to_index)
```

```python
from IPython.display import display, HTML

def print_comparison(from_index, to_index):
    display(HTML(pd.DataFrame({'content': data['Content'].values[from_index:to_index], 
                           'origin_summary': data['Summary'].values[from_index:to_index], 
                           'predict_summary': preds['Pred'].values}).to_html()))
```

```python
print_comparison(from_index, to_index)
```

```python
preds.to_csv('cache/batch_pred.csv', index=False, header=False, mode='a')
```
